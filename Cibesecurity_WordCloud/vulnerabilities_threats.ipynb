```python
#Import requests for web scraping
import pandas as pd
import requests as rq
import numpy as np
import collections
import warnings
warnings.filterwarnings('ignore')

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

import matplotlib as mpl
import matplotlib.pyplot as plt

# import k-means from clustering stage
from PIL import Image 
from matplotlib import rcParams

print('Libraries installed')
```

# First article

```python
website_url= rq.get('https://www.compuquip.com/blog/cybersecurity-threats-vulnerabilities').text 
```

```python
#Import BeautifulSoup for html structure information from our request
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')

data = [element.text for element in soup.find_all("p")] #Retrieving text from "p" founds
data = str(data) #Convert to string
data
```

```python
from stop_words import get_stop_words 
stop_words_es = get_stop_words('en') 
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

```python
stop_words = ["can", "s", "many", "vulnerabilities", "vulnerability", "cybersecurity", "security", "may"] + stop_words_es #Adding aditional StopWords
#initiate world cloud object
v_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stop_words
)

# generate the word cloud
v_wc.generate(data)
```

```python
# display the word cloud
plt.imshow(v_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
v_wc.generate(data)

# display the cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(v_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# save mask to mx_mask
mx_mask = np.array(Image.open('candado.png'))
    
print('Image downloaded and saved!')
```

```python
# instantiate a word cloud object
v_wc = WordCloud(background_color='white', max_words=2000, mask=mx_mask, stopwords=stop_words)

# generate the word cloud
v_wc.generate(data)

# display the word cloud
fig = plt.figure()
fig.set_figwidth(10) # set width
fig.set_figheight(12) # set height


plt.imshow(v_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
filtered_words = [word for word in data.split() if word not in stop_words]
counted_words = collections.Counter(filtered_words)

words = []
counts = []
for letter, count in counted_words.most_common(10):
    words.append(letter)
    counts.append(count)
```

```python
colors = cm.rainbow(np.linspace(0, 1, 10))
rcParams['figure.figsize'] = 20, 10

plt.title('Top words in the text vs their count')

plt.xlabel('Count')
plt.ylabel('Words')
plt.barh(words, counts, color=colors)
```
